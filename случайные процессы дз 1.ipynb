```python
# № 2
# генерируем выборки
exp_sample = np.random.exponential(scale = 1/2, size = 5000)
u_sample = np.random.uniform(low=0.0, high=1.0, size= 5000)

# преобразуем выборку из равномерного распределения 
transformed_sample = -np.log(1 - u_sample) / 2

# строим гистограммы
plt.figure(figsize=(10, 5))

plt.plot(1, 2, 1) 
plt.hist(exp_sample, bins = 75, color = 'blue', alpha=0.7)
plt.title("Гистограмма для выборки из экспоненциального распределения")
plt.xlabel("Значения")
plt.ylabel("Частота")
plt.show()

# Создание гистограммы для второго массива
plt.figure(figsize=(10, 5))
plt.plot(1, 2, 2) 
plt.hist(transformed_sample, bins = 75, color= 'green', alpha=0.7)
plt.title("Гистограмма для выборки из равномерного распределения при применении функции")
plt.xlabel("Значения")
plt.ylabel("Частота")
plt.show()
```

Гистограммы практически идентичны. Все дело в связи экспоненциального и равномерного распределений и обратного преобразования.
 
Экспоненциальное распределение с параметром 𝜆 описывается следующим образом:
$$
f(x) = \lambda e^{-\lambda x}, \quad x \geq 0
$$
Экспоненциальную случайную величину можно получить через преобразование равномерной случайной величины. 
𝑈 — случайная величина, равномерно распределённая на интервале 
[0,1]. Тогда случайная величина 𝑋, соответствующая экспоненциальному распределению с параметром 𝜆, выражается как:
$$
X = -\frac{\ln(1 - U)}{\lambda}
$$
В нашем случае как раз получаем при трансформации равномерной случайной величины  почти идентичное начальному экспоненциальное распределение. Возможно, что небольшие ошибки связаны с проблемой компьютеров при обработке чисел с плавающей точкой. 

P.s. использовалась статья с Википедии: https://en.wikipedia.org/wiki/Inverse_transform_sampling#Definition


```python
# №3
# создаем выборки
selection_uniform = np.random.uniform(low = 0, high= 2*np.pi, size= 5000)
selection_exp = np.random.exponential(scale= 1/1 , size = 5000)

# строим по функциям другие случайные величины
z1 = np.sqrt(2*selection_exp)*np.sin(selection_uniform)
z2 = np.sqrt(2*selection_exp)*np.cos(selection_uniform)

# Вычисление начальных моментов 3 и 4 порядка
initial_moment_3_z1 = np.mean(z1**3)  
initial_moment_4_z1 = np.mean(z1**4) 

initial_moment_3_z2 = np.mean(z2**3)  
initial_moment_4_z2 = np.mean(z2**4) 

# Вычисление центральных моментов 3 и 4 порядка
central_moment_3_z1 = sps.moment(z1, moment=3) 
central_moment_4_z1 = sps.moment(z1, moment=4) 

central_moment_3_z2 = sps.moment(z2, moment=3) 
central_moment_4_z2 = sps.moment(z2, moment=4) 

print(f"Начальный момент 3-го порядка величины z1: {initial_moment_3_z1}")
print(f"Начальный момент 4-го порядка величины z1: {initial_moment_4_z1}")
print(f"Центральный момент 3-го порядка величины z1: {central_moment_3_z1}")
print(f"Центральный момент 4-го порядка величины z1: {central_moment_4_z1}")
print('-'*145)
print(f"Начальный момент 3-го порядка величины z2: {initial_moment_3_z2}")
print(f"Начальный момент 4-го порядка величины z2: {initial_moment_4_z2}")
print(f"Центральный момент 3-го порядка величины z2: {central_moment_3_z2}")
print(f"Центральный момент 4-го порядка величины z2: {central_moment_4_z2}")

# построение гистограмм
plt.figure(figsize=(10, 5))

plt.plot(1, 2, 1) 
plt.hist(z1, bins = 75, color = 'blue', alpha=0.7)
plt.title("Гистограмма для случайной величины z1")
plt.xlabel("Значения")
plt.ylabel("Частота")
plt.show()

plt.figure(figsize=(10, 5))
plt.plot(1, 2, 2) 
plt.hist(z2, bins = 75, color= 'green', alpha=0.7)
plt.title("Гистограмма для случайной величины z2")
plt.xlabel("Значения")
plt.ylabel("Частота")
plt.show()
```

```python
# №4
# генерируем выборку 
n = 10**5
# здесь корень так как при генерации на вход должно подаваться стандартное отклонение, а не дисперсия
large_sample_norm = np.random.normal(loc = 0, scale= 1/np.sqrt(n), size= n) 
cumsum_vector = np.cumsum(large_sample_norm)

# Масштабируем вектор накопленных сумм для получения правильных значений W_i/n
scaled_cumsum = cumsum_vector / np.sqrt(n)

# Создание графика
t = np.linspace(0, 1, n) 

plt.plot(t, scaled_cumsum, label="Траектория процесса")
plt.title("Приближенная траектория винеровского процесса")
plt.xlabel("t")
plt.ylabel("W(t)")
plt.grid(True)
plt.legend()
plt.show()
```

Здесь я попробовала реализовать следующий алгоритм: выбирается размер выборки n (в моем случае -  100 000) и счетчик "четности". Далее создается цикл, который повторяется n раз, каждый раз генерируя новые значения для переменных x,y и вычисляя z, являющеся отношением x к y. Затем я округляю по математическому правилу z до целого числа и проверяю его на четность. В случае четности увеличиваю счетчик на единицу. После выхода из цикла имеем размер выборки и кол-во среди всех значений четных. Делим последнее на размер выборки - получаем итоговую вероятность, которую затем округляем до 2-х знаков после запятой.

```python
# №5
n = 100000
even_count = 0
for _ in range(n):
    x = np.random.uniform(low=0.0, high=1.0, size= 1)
    y = np.random.uniform(low=0.0, high=1.0, size= 1)
    z = np.round(x / y)
    # Проверка на четность
    if z % 2 == 0:
        even_count += 1

probability_even = even_count / n

print(f'Приближенное значение вероятности того, что величина x/y ближе к четному числу, чем к нечетному: {np.round(probability_even,2)}')


```

Решение: выбираем объем выборки n. Далее генерируем выборки этого объема из распределения Дирихле (выбрала его, так как элементы каждой выборки в сумме дают единицу - "длина" нашей палки. Как вариант можно было использовать равномерное распределение с нормализацией). Таких выборок я решила сгенерировать 100 000. Далее в каждой выборке проходим по всевозможным комбинациям (на этом моменте запускаем счетчик всевозможных комбинаций) из трех элементов каждую такую комбинацию проверяя на неравенство треугольника. Если треугольник существует, то запускается еще один счетчик, который показывает количество "успешных" комбинаций, которые удовлетворяют неравенству треугольника возможность составить треугольник. Делим один счетчик на другой - получаем нужную вероятность.

```python
# №6

# Размер выборки
#n = 5
#n = 15
n = 30

# Генерация выборок с распределением Дирихле
sample_dirichlet = np.random.dirichlet(np.ones(n), size=100000)  # 100 тысяч выборок по n элементов

# Счётчик для треугольных комбинаций
triangle_count = 0

# Общее количество проверенных комбинаций
total_combinations = 0

# Функция для проверки неравенства треугольника
def is_triangle(a, b, c):
    return (a + b > c) and (a + c > b) and (b + c > a)

# Проход по всем выборкам
for sample in sample_dirichlet:
    # Перебор всех комбинаций по 3 элемента
    for combination in itertools.combinations(sample, 3):
        total_combinations += 1
        # Проверяем, можно ли составить треугольник
        if is_triangle(*combination):
            triangle_count += 1

# Вычисление вероятности
probability = triangle_count / total_combinations

print(f"Вероятность того, что из трёх элементов можно составить треугольник: {probability:.2f}")

```
